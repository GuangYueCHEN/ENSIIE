```python
%matplotlib inline
import pandas as pd
import numpy  
import matplotlib.pylab as plt

import os
###############################################################
### statistiques descriptives 
def Stats_des(data):
    Stats_t=['nb données','min','max','moyenne','variance']
    Calculs=[0]*len(Stats_t)
    Calculs[0]=int("{:d}".format(len(data)))
    Calculs[1]=float("{:.4f}".format(numpy.min(data)))
    Calculs[2]=float("{:.4f}".format(numpy.max(data)))
    Calculs[3]=float("{:.4f}".format(numpy.mean(data)))
    Calculs[4]=float("{:.4f}".format(numpy.var(data)))
    
    ## création d'un dictionnaire
    Dic_Stats={}
    for i in range(len(Stats_t)):
        Dic_Stats[Stats_t[i]]=Calculs[i]
    
    return Dic_Stats
####################################################################

print("---"*35)
for i in range(2,len(Liste_donnees)): 
    F='./DonCAC40//'+Liste_donnees[i]
    data = pd.read_csv(F, delimiter=',')
   
    data=data.dropna()  # supprimer les données manquantes

    D=Stats_des(data[data.columns[5]])
    
    print("données de : %s \n"%Liste_donnees[i])
    print(D)
    print("---"*35)
```

```python
def Model_AR1(data):
    alpha1 = sum([data[t-1]*data[t-2] for t in range(2,len(data))])/sum([data[t-1]**2 for t in range(2,len(data))])
    return alpha1
```

```python
def Modele_AR_general(data,p):
    
    M = numpy.zeros((p,p))
    for i in range( p):
        for j in range(p):
            M[i,j] = sum([data[t-1-i]*data[t-1-j] for t in range(p, len(data))])
            
    B = numpy.zeros((p,1))
    for i in range(0, p):
        B[i,0] = sum([data[t]*data[t-1-i] for t in range(p, len(data))])
    alpha = numpy.zeros((p,1))
    alpha = numpy.matmul(numpy.linalg.inv(M) , B)
    
    Y=[]
    for t in range(p,len(data)):
        Y.append(numpy.sum([alpha[k-1]*data[t-k] for k in range(1,p+1)]))
    plt.plot(Y,':r')
    plt.plot(data[p:])
    plt.show()
    return alpha
```

```python
def ARp_simul(T,coeff,sigma=1.54):
    p=len(coeff)
    e=numpy.random.normal(0,sigma,T)
    data=[0]*T
    data[0]=e[0]
    for t in range(1,T):
        if t<p:
            data[t]=numpy.sum([coeff[k]*data[t-k] for k in range(t)] + e[t])
        else:
            data[t]=numpy.sum([coeff[k-1]*data[t-p+k] for k in range(1,p+1)]+e[t])
    return data
```

# fonction d'autocorrélation 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy
import pandas as pd
import os

##########################################################################
## fonction d'autocorrélation (l'équation 33 du cours avec h=0, page 19)
#自相关
def TM_ACF(data,lag=1):
    data=numpy.array(data,dtype=numpy.float64)
    m=numpy.mean(data)
    T=len(data)
    V=sum([(data[t]-m)**2 for t in range(T)])
    #print("m=%s \t v=%s"%(m,V))
    C=sum([(data[t]-m)*(data[t+lag]-m) for t in range(T-lag)])
    R=C/(1.0*V)
    return R
```

```python
data = ARp_simul(100, [0.258, -0.858, 0.12, 0.11])
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111)
# cacher axes  'haut' et 'droit':
for side in ['right','top']:
    ax.spines[side].set_visible(False)
    
k = numpy.arange(1,20,1) 
ACF=[TM_ACF(data[25:],lag=j) for j in k]





#制作垂直线
markerline, stemlines, baseline = plt.stem(k, ACF, '-.')
plt.setp(baseline, 'color', 'r', 'linewidth', 1)
    
#plt.title("données de : %s"%Liste_donnees[i])
plt.show()
```
